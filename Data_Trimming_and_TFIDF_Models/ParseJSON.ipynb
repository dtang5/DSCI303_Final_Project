```
root_path = '/content/drive/My Drive/DSCI 303 Final Project/Data_Trimming_and_TFIDF_Models/'
```

```
import json
read_filename = root_path + "review500k.json"
data = {}
data['Review:Rating'] = []
with open(read_filename, 'r') as f:
    for line in f:
        d = json.loads(line)
        data['Review:Rating'].append({
            'review': d['text'], 
            'stars': d['stars'],
        })    
    f.close()
```

```
with open(root_path + "trimmed_review_500k.json", 'w') as outfile:
    json.dump(data, outfile)
```
