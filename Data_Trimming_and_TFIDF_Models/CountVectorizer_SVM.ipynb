```
with open('/content/drive/My Drive/DSCI 303 Final Project/Data_Trimming/preprocessed_150k_reviews.json') as json_file2: # Open json file to load
    data2 = json.load(json_file2)
    data3 = json.loads(data2)
```

```
df = pd.DataFrame.from_dict(data3)
X_raw = df['review']
Y_raw = df['stars']
```

```
from sklearn.feature_extraction.text import CountVectorizer
```

```
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(X_raw)
```

```
X
```

```
Y = Y_raw
```

```
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, accuracy_score
# from sklearn.model_selection import GridSearchCV
```

```
from sklearn.model_selection import cross_val_score
```

```
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=101)
```

```
from sklearn import linear_model
```

```
# L2
svm = linear_model.SGDClassifier(max_iter=1000, tol=1e-3) # L2 penalty
svm_cv_scores = cross_val_score(svm, x_train, y_train, cv=8)
print(svm_cv_scores)
```

```
np.average(np.array([0.64015732,0.6426,0.63726667,0.64553333,0.638,0.6404,0.64226667,0.64424295]))
```

```
# L1
svm = linear_model.SGDClassifier(max_iter=1000, tol=1e-3, penalty='l1') # L1 penalty
svm_cv_scores = cross_val_score(svm, x_train, y_train, cv=8)
print(svm_cv_scores)
```

```
np.average(np.array([0.6130258,0.61826667,0.6214,0.6264,0.62173333,0.6192,0.6182,0.62224148]))
```

```
# elasticnet
svm = linear_model.SGDClassifier(max_iter=1000, tol=1e-3, penalty='elasticnet') # elasticnet penalty
svm_cv_scores = cross_val_score(svm, x_train, y_train, cv=8)
print(svm_cv_scores)
```

```
np.average(np.array([0.64255716,0.64353333,0.63613333,0.64826667,0.64566667,0.6394,0.646,0.64397627]))
```

```
# Support Vector Machine final
svm = linear_model.SGDClassifier(max_iter=1000, tol=1e-3, penalty='elasticnet')
svm.fit(x_train,y_train)
predsvm = svm.predict(x_test)
print("Confusion Matrix for Support Vector Machines:")
print(confusion_matrix(y_test,predsvm))
print("Score:",round(accuracy_score(y_test,predsvm)*100,2))
print("Classification Report:",classification_report(y_test,predsvm))
```
