```
import json
read_filename = root_path + "review150k.json"
data = {}
data['Review:Rating'] = []
with open(read_filename, 'r') as f:
    for line in f:
        d = json.loads(line)
        data['Review:Rating'].append({
            'review': d['text'], 
            'stars': d['stars'],
            'useful': d['useful'],
            'funny' : d['funny'],
            'cool' : d['cool']
        })    
    f.close()
```

```
root_path = '/content/drive/My Drive/DSCI 303 Final Project/Data Trimming/'
```

```
with open(root_path + "trimmed_review_with_reactions150k.json", 'w') as outfile:
    json.dump(data, outfile)
```
